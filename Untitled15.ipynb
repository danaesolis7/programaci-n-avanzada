```python
import numpy as np
import matplotlib.pyplot as plt

# DATOS
m = float(input("Masa (kg): "))
theta = float(input("Ángulo (grados): "))
mu = float(input("Coeficiente de fricción: "))

g = 9.81

# CONVERSIONES
theta_rad = np.radians(theta)

# CÁLCULOS
Fx = m * g * np.sin(theta_rad)
N = m * g * np.cos(theta_rad)
Ff = mu * N
Fnet = Fx - Ff
a = Fnet / m

# RESULTADOS
print("\nRESULTADOS")
print(f"Fuerza paralela: {Fx:.2f} N")
print(f"Fuerza normal: {N:.2f} N")
print(f"Fuerza de fricción: {Ff:.2f} N")
print(f"Aceleración: {a:.2f} m/s²")

# GRÁFICA
t = np.linspace(0,10,100)
v = a*t

plt.plot(t,v)
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.title("Velocidad vs Tiempo")
plt.grid()
plt.show()
```
