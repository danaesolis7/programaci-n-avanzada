```python
import numpy as np
import matplotlib.pyplot as plt

# DATOS
m = float(input("Masa del robot (kg): "))
vi = float(input("Velocidad inicial (m/s): "))
vf = float(input("Velocidad final (m/s): "))
t = float(input("Tiempo (s): "))

# CÁLCULOS
a = (vf - vi)/t
F = m*a

print("\nRESULTADOS")
print(f"Aceleración: {a:.2f} m/s²")
print(f"Fuerza necesaria: {F:.2f} N")

# GRÁFICA
tiempo = np.linspace(0,t,100)
velocidad = vi + a*tiempo

plt.plot(tiempo,velocidad)
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.title("Movimiento del Robot")
plt.grid()
plt.show()
```
