```python
import numpy as np
import matplotlib.pyplot as plt

# DATOS
k = float(input("Constante del resorte k (N/m): "))
m = float(input("Masa (kg): "))
A = float(input("Amplitud (m): "))

# CÁLCULOS
w = np.sqrt(k/m)
T = 2*np.pi/w
f = 1/T

print("\nRESULTADOS")
print(f"Frecuencia angular: {w:.2f} rad/s")
print(f"Periodo: {T:.2f} s")
print(f"Frecuencia: {f:.2f} Hz")

# MOVIMIENTO
t = np.linspace(0,10,1000)
x = A*np.cos(w*t)

# GRÁFICA
plt.plot(t,x)
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.title("Oscilador Armónico")
plt.grid()
plt.show()
```
